```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from forecaster import Forecaster

from sklearn.linear_model import Ridge, LinearRegression
from xgboost import XGBRegressor
```

```python
df = pd.read_csv("sales-of-shampoo-over-a-three-ye.csv")
df.columns = ['Date', 'Shampoo']
```

```python
df['Date'] = pd.to_datetime(df['Date'].apply(lambda s: '200'+s))
```

```python
df.plot(x='Date', y='Shampoo', figsize=(15, 3))
```

```python
model = Forecaster(
    df,
    x = 'Date',
    y = 'Shampoo'
)
```

```python
pred, scores = model.validate(Ridge, seasonality = True, lag = True, plot = True)
```
