```python
import pandas as pd
from forecaster import Forecaster
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/main/examples/example_wp_log_peyton_manning.csv')
df.head()
```

```python
model = Forecaster(
    df,
    x = 'ds',
    y = 'y'
)
```

```python
model.make_future_dataframe(periods=365)
model.df.tail()
```

```python
from sklearn.linear_model import Ridge
y_valid, scores = model.validate(Ridge, seasonality = True, lag = True, plot = True)
```

```python
forecast, scores = model.forecast(Ridge, plot = True)
```

```python
start_date = (df['ds']).min()
end_date = (df['ds']).max()
missing_dates = pd.date_range(start_date, end_date).difference((df['ds']).unique())
```

```python
for missing_date in missing_dates:
    view = df[df['ds'] == df['ds'].iloc[0]]
    view['y'] = 0
    view['ds'] = missing_date
    df = pd.concat([df, view])
```
